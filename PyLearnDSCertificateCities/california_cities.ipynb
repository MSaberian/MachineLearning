```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn
```

```python
data = pd.read_csv('input/california_cities.csv')
data.head()
```

```python
seaborn.set()
plt.scatter(data['latd'], data['longd'], s=data['area_land_km2'], alpha=0.5, linewidths=0,
             c=np.log10(data['population_total']),
             cmap='gist_ncar')
             
plt.clim(3, 7)
plt.gca().set_aspect('equal')
plt.colorbar(label='log10 population')
plt.xlabel('Latitute')
plt.ylabel('Longtitude')
plt.title('Area and population of Califurnia Cities')
for area in [100, 300, 500]:
    plt.scatter([], [], c='k', alpha=0.3, s=area, label=str(area) + 'km$^2$')
plt.legend(scatterpoints=1, frameon=False, labelspacing=1, title='City Areas')
plt.show()
```
