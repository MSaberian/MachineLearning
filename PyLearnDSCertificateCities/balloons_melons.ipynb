```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig = plt.figure()
ax = fig.add_subplot(projection='3d')

n = 200
std = 2

xs = np.random.normal(15, std, n)
ys = np.random.normal(25, std, n)
zs = np.random.normal(.05, std, n)
ax.scatter(xs, ys, zs, marker='o')

xs = np.random.normal(12.5, std, n)
ys = np.random.normal(12.5, std, n)
zs = np.random.normal(10, std, n)
ax.scatter(xs, ys, zs, marker='o')

ax.set_xlabel('width (cm)')
ax.set_ylabel('length (cm)')
ax.set_zlabel('weight (kg)')

ax.view_init(20,10)

plt.title('balloons and melons')
plt.show()
plt.savefig('sdfsdf.jpg')
```
