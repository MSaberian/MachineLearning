```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('input/california_cities.csv')
data.head()
```

```python
plt.scatter(data['latd'], data['longd'], s=data['area_land_km2'], alpha=0.5, linewidths=0,
             c=np.log10(data['population_total']),
             cmap='gist_ncar')
plt.colorbar(label='log10 population')
plt.xlabel('Latitute')
plt.ylabel('Longtitude')
plt.title('Area and population of Califurnia Cities')
plt.show()
```
