```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('input/COVID-19.csv')

data.head(10)
```

```python
data4 = data.query("Date>='2020-04-01' and Date<'2020-05-01'").reset_index(drop=True)

data4.head(-10)
```

```python
data4c = data4.groupby('CountryRegion')['Active'].max().reset_index()
data4cs = data4c.sort_values(by="Active", ascending=False).reset_index()


data4cs.head(8)
```

##  death rate in Iran on different days

```python
data_iran = data.query("CountryRegion == 'Iran'").reset_index(drop=True)
data_iran['Date'] = pd.to_datetime(data_iran['Date'],yearfirst=True)    
data_iran['New_Date'] = data_iran['Date'].map(lambda date:f'{date.year}/{date.month}/{date.day}')
data_iran.head(-10)
```

```python
plt.figure(figsize=(12,6))
plt.plot(data_iran['New_Date'].to_numpy(dtype=str), np.log10(data_iran['New cases'].to_numpy()), color="skyblue")
plt.plot(data_iran['New_Date'].to_numpy(dtype=str), np.log10(data_iran['New deaths'].to_numpy()), color="lightpink")

plt.fill_between(data_iran['New_Date'].to_numpy(dtype=str), np.log10(data_iran['New cases'].to_numpy()), color="skyblue", alpha=0.5, label='New cases')
plt.fill_between(data_iran['New_Date'].to_numpy(dtype=str), np.log10(data_iran['New deaths'].to_numpy()), color="lightpink", alpha=0.5, label='New deaths')

plt.title('New cases and new death in Iran')
plt.xlabel('Date')
plt.ylabel('Number (log10)')
plt.xticks(rotation='vertical')
plt.xticks(np.arange(0, len(data_iran), 20))
plt.ylim(bottom = 0) 
plt.legend(['New cases','New deaths'])
plt.show()
```
